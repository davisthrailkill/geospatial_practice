```python
# import packages
from shapely.geometry import Point
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import folium

%matplotlib inline
```

```python
zip_codes = gpd.read_file('data/zip_codes.geojson')
zip_codes.head()
```

```python
mta_stops = pd.read_csv('data/mta_stops.csv')
mta_stops.head()
```

```python
mta_stops = mta_stops.rename(columns={'Stop ID Number':'Stop_ID','Stop Abbreviation':'Stop_Abbreviation','StopName':'Stop_Name','Route Number':'Route_Number','Route Name':'Route_Name','Mapped Location':'Mapped_Loc'})
mta_stops.head()
```

```python
public_health = pd.read_csv('data/public_health_loc.csv')
public_health
```

```python
public_health['Location 2'] = public_health['Location 1'].apply(lambda x: x.split('\n'))
public_health
```

```python
public_health['Location 3'] = [loc[2] for loc in public_health['Location 2']]
```

```python
public_health['Location 3'] = public_health['Location 3'].str.replace('(','')
```

```python
public_health['Location 3'] = public_health['Location 3'].str.replace(')','')
```

```python
public_health['Location 4'] = public_health['Location 3'].str.split(',')
```

```python
public_health['Lng'] = [loc[1] for loc in public_health['Location 4']]
```

```python
public_health['Lat'] = [loc[0] for loc in public_health['Location 4']]
```

```python
public_health
```

```python
public_health = public_health.drop(columns=['Location 2', 'Location 3'])
public_health
```

```python
public_health = public_health.rename(columns={'Clinic Name':'Clinic','Location 1':'Address','Location 4':'Loc'})
public_health
```

```python
public_health['Lng'] = public_health['Lng'].astype(float)
public_health['Lat'] = public_health['Lat'].astype(float)
```

```python
mta_stops['Mapped_Loc'] = mta_stops.apply(lambda x: Point((float(x.Longitude), float(x.Latitude))), axis=1)
```

```python
mta_stops_geo = gpd.GeoDataFrame(mta_stops, crs = zip_codes.crs, geometry = mta_stops.Mapped_Loc)
```

```python
mta_stops_geo.head()
```

```python
bus_stops = gpd.sjoin(mta_stops_geo, zip_codes, op = 'within')
bus_stops.head()
```

```python
public_health['Geom'] = public_health.apply(lambda x: Point((float(x.Lng), float(x.Lat))), axis=1)
```

```python
public_health_geo = gpd.GeoDataFrame(public_health, crs = zip_codes.crs, geometry = public_health.Geom)
```

```python
public_health_geo.head()
```

```python
pub_health = gpd.sjoin(public_health_geo, zip_codes, op = 'within')
pub_health.head()
```

```python
downtown_pub_health = pub_health.loc[pub_health['zip'] == '37203']
downtown_pub_health
```

```python
downtown_bus_stops = bus_stops.loc[bus_stops['zip'] == '37203']
downtown_bus_stops.head()
```

```python
downtown_zip_polygon = zip_codes.loc[zip_codes['zip'] == '37203']
downtown_zip_polygon.head()
```

```python
center = downtown_zip_polygon.geometry.centroid
center
```

```python
# get and store the first occurence
center_point = center.iloc[0]
```

```python
# reverse the order for folium, since it wants latitude first and shapely Points have longitude first
urban_center = [center_point.y, center_point.x]
urban_center
```

```python
nashville_map = folium.Map(location = urban_center, zoom_start = 13.8)
nashville_map
```

```python
for row in downtown_bus_stops.iterrows():
    row_values = row[1]
    print(row_values)
```

```python
for row in downtown_pub_health.iterrows():
    row_values = row[1]
    print(row_values)
```

```python
folium.GeoJson(downtown_zip_polygon).add_to(nashville_map)

for row in downtown_pub_health.iterrows():
    row_values = row[1]
    location = [row_values['Lat'], row_values['Lng']]
    popup = (str(row_values['Clinic']) + '<br/>' +
            'Hours: ' + str(row_values['Hours']) + '<br/>' +
            'Phone: ' + str(row_values['Phone'])).replace("'", "`")
    marker = folium.Marker(location = location, popup = popup, icon=folium.Icon(color='red', icon='fa-hospital-o', 
                                                                                icon_color='white', prefix='fa'))
    marker.add_to(nashville_map)

# display map
nashville_map
```

```python
# iterate through bus_stops and create locations and markers
# note that latitude must come first for folium map to work

for row in downtown_bus_stops.iterrows():
    row_values = row[1]
    location = [row_values['Latitude'], row_values['Longitude']]
    popup = (str(row_values['Stop_ID']) + ': ' +
             str(row_values['Stop_Abbreviation']) + '<br/>' +
             'Stop Name: ' + str(row_values['Stop_Name']) + '<br/>' +
             'Route Name: ' + str(row_values['Route_Name'])).replace("'", "`")
    marker = folium.Marker(location = location, popup = popup, icon=folium.Icon(color='blue', icon='bus',
                                                                               icon_color='white', prefix='fa'))
    
    marker.add_to(nashville_map)

# display map
nashville_map
```
